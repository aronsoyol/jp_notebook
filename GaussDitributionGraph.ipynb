```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import math
from __future__ import unicode_literals
```

```python
x = np.arange(-3, 3, 0.15)
y = np.arange(-3, 3, 0.15)
```

```python
X, Y = np.meshgrid(x, y)
```

```python
ax_share = None
variable =[0.1,0.2,0.4,0.6,0.8,0.9]
figures=[]
for i, s in enumerate(variable):
    s2=s
    ro=0.1
    
    Q=1/(1-ro**2)*(X**2+2*ro*X*Y/s2+(Y/s2)**2)
    Z = 1/(2.0* np.pi*s2 *(math.sqrt(1.0-ro**2))) * np.exp(-0.5*Q)
           
    fig = plt.figure(figsize=(20,20))
    
    ax = None

    if i ==0:
        ax = fig.add_subplot(1, 1, 1, projection='3d')
        ax_share = ax
    else:
        ax = fig.add_subplot(1, 1,1, projection='3d', sharex=ax_share, sharey=ax_share, sharez=ax_share)
        
        
#     ax = Axes3D(ax1)
    ax.plot_surface(X, Y, Z, rstride=1, cstride=1,color = '0.75', linewidth=0.2,alpha=0.7)
#     ax.annotate("$\\sigma_2=%0.2f$"%(s2), xy=(0.3 ,0.5), xycoords='axes fraction', fontsize=15,horizontalalignment='right')
#     fig.savefig("1%d.pdf"%(i))
    figures.append(fig)
for i, r in enumerate(variable):
    s2=1
    ro=r
    
    Q=1/(1-ro**2)*(X**2+2*ro*X*Y/s2+(Y/s2)**2)
    Z = 1/(2* np.pi*s2 *((1-ro**2)**0.5)) * np.exp(-0.5*Q)
    
    fig = plt.figure(figsize=(20,20))

    ax = fig.add_subplot(1, 1, 1, projection='3d',  sharex=ax_share, sharey=ax_share, sharez=ax_share)
#     ax = Axes3D(ax1)
    ax.plot_surface(X, Y, Z, rstride=1, cstride=1,color = '0.75', linewidth=0.2,alpha=0.7)
#     ax.set_title()
#     ax.annotate("$\\rho=%0.2f$"%(ro), xy=(0.3, 0.5), xycoords='axes fraction', fontsize=15,horizontalalignment='right')
#     fig.savefig("2%d.pdf"%(i))
    figures.append(fig)
    
plt.show()
```

```python
for i, fig in enumerate(figures):
#     bbox_inches='tight'余計な余白を取り除く
    fig.savefig("%d-%d.pdf"%(i/6,i %6),bbox_inches='tight')
```
